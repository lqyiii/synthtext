```python
# read corpus

import random

file_list=os.listdir('/home/ubuntu/liqiyi/SynthText_Chinese_version/data/newsgroup/')
first_round_sentence_list=[]
for filename in file_list:
    try:
        rf=codecs.open('/home/ubuntu/liqiyi/SynthText_Chinese_version/data/newsgroup/'+filename,'r','utf-8')
        while True:
            line_str=rf.readline().strip()
            if len(line_str)==0:
                break
            first_round_sentence_list.append(line_str)

        rf.close()
    except:
        rf.close()
        print filename.decode('utf-8')
    
first_round_sentence_list=list(set(first_round_sentence_list))
print len(first_round_sentence_list)
random.shuffle(first_round_sentence_list)
len(first_round_sentence_list)


second_round_sentence_list=[]
file_list=os.listdir('/home/ubuntu/liqiyi/SynthText_Chinese_version/data/data_corpus/')
print 'file count:',len(file_list)

for filename in file_list:
    try:
        rf=codecs.open('/home/ubuntu/liqiyi/SynthText_Chinese_version/data/data_corpus/'+filename,'r','utf-8')
        while True:
            line_str=rf.readline().strip()
            if len(line_str)==0:
                break
            second_round_sentence_list.append(line_str)

        rf.close()
    except:
        rf.close()
        print filename.decode('utf-8')
    
second_round_sentence_list=list(set(second_round_sentence_list))
print len(second_round_sentence_list)
random.shuffle(second_round_sentence_list)


third_round_sentence_list=[]

file_list=os.listdir('/home/ubuntu/liqiyi/SynthText_Chinese_version/data/sogou_corpus/')
print 'file count:',len(file_list)

for filename in file_list:
    try:
        rf=codecs.open('/home/ubuntu/liqiyi/SynthText_Chinese_version/data/sogou_corpus/'+filename,'r','gbk')
        while True:
            line_str=rf.readline().strip().split(u'	')[0]
            line_str=line_str.replace(u'-',u'')
            if len(line_str)==0:
                break
            third_round_sentence_list.append(line_str)

        rf.close()
    except:
        rf.close()
        print filename.decode('utf-8')
    
third_round_sentence_list=list(set(third_round_sentence_list))
print len(third_round_sentence_list)
random.shuffle(third_round_sentence_list)
len(third_round_sentence_list)



news_list=[]

sogou_news_list=os.listdir('/home/ubuntu/liqiyi/SynthText_Chinese_version/data/sogou_news/SogouCA/')
for filename in sogou_news_list:
    rf=open('/home/ubuntu/liqiyi/SynthText_Chinese_version/data/sogou_news/SogouCA/'+filename,'r')
    print filename
    
    while True:
        str_line=rf.readline().strip()
        if len(str_line)==0:
            break
            
        try:
            utf_str=str_line.decode('gbk','ignore')
        except:
            utf_str=str_line.decode('gb2312')
        
        if '<contenttitle>' in utf_str:
            start_pos=utf_str.find('<contenttitle>')
            end_pos=utf_str.find('</contenttitle>')
            text=utf_str[start_pos+len('<contenttitle>'):end_pos]
            if len(text)>0:
                news_list.append(text)
            
        if '<content>' in utf_str:
            start_pos=utf_str.find('<content>')
            end_pos=utf_str.find('</content>')
            text=utf_str[start_pos+len('<content>'):end_pos]
            if len(text)>0:
                news_list.append(text)
            
    rf.close()



text_list=first_round_sentence_list+second_round_sentence_list+third_round_sentence_list+news_list
print len(text_list)
```

```python
def select_samples_new(word_sentence_dict, min_balance_thresh=300,freq_del_thresh=400, max_count=500):
    global text_list
    # 从text_list中抽样
    find_dict={word:1 for word in word_sentence_dict.keys() if word_sentence_dict[word]['count']<min_balance_thresh}
    print 'find word count:',len(find_dict)

    for sentence in text_list:
        # 针对每个字
        for pos in range(len(sentence)):
            word=sentence[pos]
            if find_dict.has_key(word):
            #if word_sentence_dict.has_key(word):
                if word_sentence_dict[word]['count']<min_balance_thresh:
                    cntSum_dict={}
                    for i in range(5):
                        # 选择最优的 序列
                        part_str=sentence[pos-5+i:pos+i]
#                         sentece_list_tmp=word_sentence_dict[word]['sentence_list']
#                         cnt_before=len(sentece_list_tmp)
#                         cnt_after=len(set(sentece_list_tmp+[part_str]))
                        if len(part_str)>=2:# 查看长度是否满足要求，应该>2  且不重复？
                            # 检查单词是否都在词典中
                            count=[word_sentence_dict.has_key(j) for j in part_str].count(False)
                            if count==0:
                                # 查看各自单词的计数
                                array_tmp=np.array([word_sentence_dict[j]['count'] for j in part_str])
                                over_cnt=array_tmp[array_tmp>max_count].size
                                if over_cnt==0:
                                    cnt_sum=sum([word_sentence_dict[w]['count'] for w in part_str])/float(len(part_str))
                                    cntSum_dict[cnt_sum]=part_str

                    cntSum_list=cntSum_dict.keys()
                    if len(cntSum_list)>0:
                        min_cntSum=min(cntSum_list)
                        str_tmp=cntSum_dict[min_cntSum]

                        word_sentence_dict[word]['sentence_list'].append(str_tmp)

                        for w in str_tmp:
                            word_sentence_dict[w]['count']+=1
        
        
        rand_int=random.randint(0,10)
        if rand_int>=8: # 删除掉 高频词句
            del_word_dict={word:1 for word in word_sentence_dict.keys() if word_sentence_dict[word]['count']>freq_del_thresh}
            for word in del_word_dict:
                sentence_list_tmp=word_sentence_dict[word]['sentence_list']

                # 计算每个句子中word的cnt加和
                cntSum_del_dict={}
                for sen_del in sentence_list_tmp:
                    cntList_tmp=[word_sentence_dict[w]['count'] for w in sen_del]
                    if min(cntList_tmp)>freq_del_thresh:
                        cntSum=sum(cntList_tmp)/float(len(sen_del))
                        if cntSum_del_dict.has_key(cntSum):
                            cntSum_del_dict[cntSum].append(sen_del)
                        else:
                            cntSum_del_dict[cntSum]=[sen_del]

                key_list=cntSum_del_dict.keys()
                if len(key_list)>0:
                    # 去掉20%高频语句
                    key_list=sorted(key_list,reverse=True)
                    key_chosed=key_list[:int(np.ceil(0.2*len(key_list)))]
                    sen_del_list=[]
                    for key_t in key_chosed:
                        sen_del_list+=cntSum_del_dict[key_t]

                    # 直接用set相减
                    sentence_list_new=list(set(word_sentence_dict[word]['sentence_list'])-set(sen_del_list))
                    word_sentence_dict[word]['sentence_list']=sentence_list_new

                    for sen_del in sen_del_list:
                        #word_sentence_dict[word]['sentence_list'].remove(sen_del)
                        for w in sen_del:
                            word_sentence_dict[w]['count']-=1
                        
                        rand_num=random.randint(0,100000)
                        if rand_num>=100000-1000:
                            print len(sen_del_list)
                            print 'remove',sen_del
                        
                        if rand_num>=100000-100:
                            import time
                            time_str=time.strftime('%Y%m%d%H%M',time.localtime())
                            pkl.dump(word_sentence_dict,open('./saved_file/word_sentence_dict_'+time_str,'wb'))

    # 计算cnt，若满足要求，那么便结束递归
    word_cnt_list=sorted([word_sentence_dict[key]['count'] for key in word_sentence_dict.keys()])
    print word_cnt_list[:5]
    if word_cnt_list[0]>=min_balance_thresh:
        return word_sentence_dict
    else:
        return select_samples_new(word_sentence_dict)
```
